```python
import pandas as pd
from matplotlib import pyplot as mpl
#importando bibliotecas
```

```python
#lendo o csv e armazendo-o em uma variável:
vacinas = pd.read_csv("SEUARQUIVO.csv", sep=";", encoding="ISO-8859-1",low_memory=False)
```

```python
vacinas.head() #exibindo o início da lista do pandas
```

```python
vacinas.columns.values #vai mostrar o título de todas as colunas do seu gráfico
```

```python
filtroVacina = ['paciente_enumSexoBiologico']
```

```python
novoGrafico = vacinas.filter(items=filtroVacina) #filtrando
```

```python
novoGrafico
```

```python
geraGraf = novoGrafico.value_counts().sort_index()
```

```python
#configurando o matplot
labels = ['Feminino', 'Masculino']
colors = ['#F2DD0C', '#192DFF']
mpl.pie(geraGraf, labels=labels,colors=colors, wedgeprops = {'edgecolor':'black'})
mpl.title("Vacinação entre sexos")
```
